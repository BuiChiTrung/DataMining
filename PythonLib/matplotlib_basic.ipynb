```python
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
import numpy as np 
import pandas as pd 

```

## Two interface for the price of one 

### MATLAB-style interface
Vẽ second panel rồi thì ko quản lí đc first panel

```python
# create a plot figure can config figure size, border, background color v.v
plt.figure(figsize = [6.4, 4.8])

# trả về số điểm cách đều trên một đoạn nào đó, default = 50 
x = np.linspace(0, 10, 30)

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '--', color = "red")
```

### Object-oriented interface
1. Figure: (an instance of the class plt.Figure ) can be thought of as a
single container that contains all the objects representing axes, graphics, text, and
labels. It is the topmost layer of the plot (kind of big-picture)

2. Axes: The axes (an instance of the class plt.Axes ) is what we see above: a bounding
box with ticks and labels, which will eventually contain the plot elements that make
up our visualization

```python
# chia fingure thành 1x2 subplots
fig, axe = plt.subplots(1,2, figsize = (6,4))

# có thể edit bất kỳ subplot nào bằng axe[i]
# chú ý nến subplots(2, 2) thì các axes là [0, 0] [0, 1] [1, 0] [1, 1]
axe[0].text(0.5,0.5,'(1,2,1) Using Axes',ha = 'center', fontsize = 15)
axe[1].text(0.5,0.5,'(1,2,2) Using Axes',ha = 'center', fontsize = 15)
plt.show()
```

```python
baseball = pd.read_csv('MikeTroutData.csv')

# Chuyển thông tin mỗi cột thành 1 Serie
year = baseball['Year']
hits = baseball['H']
at_bats = baseball['AB']
home_runs = baseball['HR']
salary = baseball['Salary']

baseball
```

## Bar plot

### Vertical

```python

# graph name
plt.suptitle('Hit each year')

plt.bar(year, hits)

# axis' names
plt.xlabel('Year')
plt.ylabel('#Of hit')

plt.show()
```

### Horizontal

```python
plt.barh(year, hits, color = "red")

plt.xlabel('#Of hit')
plt.ylabel('Year')

plt.show()
```

## Combine plots

```python
plt.suptitle("A player At bat and Hit per year")

plt.xlabel('Year')

# Add label to use legend
plt.bar(year, hits, label = 'Hit')
plt.plot(year, at_bats, color = "yellow", label = 'At bat')

# Chú thích đồ thị 
plt.legend()

plt.show()

```

## Stacked bar chart


```python
plt.suptitle('Home run per Hit')

plt.xlabel('Year')

plt.bar(year, hits, label = 'Hit')
plt.bar(year, home_runs, color = "black", label = 'Home run')

plt.legend()
plt.show()
```

## Grouped Bar Chart


```python
plt.xlabel('Year')
plt.suptitle('Mike Trout - Home Runs vs Total Hits')

plt.xticks(rotation=45)         #rotates labels by 45 degrees
plt.xticks(year)                #shows all years in label

plt.bar(year, hits, width=.2, label='Hits')
plt.bar(year+0.2, home_runs, width=.2, label='Home Runs')        #moved the bars around manually
plt.legend()
plt.show()
```

```python
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.xticks(year)                #shows all years in label

plt.ylabel('# of Hits')           
plt.suptitle('Mike Trout Hits per year')

for bar in plt.bar(year, hits):        
    plt.text(bar.get_x() + 0.4,             # x position of label
             bar.get_height() - 20,         # y position of label
             bar.get_height(),              # actual value of label
             ha='center',                     # căn lề
             color='black')
    print(bar.get_x(), end = ' ')
```

```python
## Math with DataFrame

plt.suptitle('Cost per Home run')

plt.xlabel('Year')
plt.xticks(year)
plt.xticks(rotation=45)

plt.ylabel('Price')

# new serie
cost_per_home_run = salary / home_runs

plt.bar(year, cost_per_home_run)

plt.show()
```

## Scatter Plot

### plot vs scatter
1. scatter cho phép tùy chỉnh từng điểm dữ liệu, plot thì các điểm  đc plot giống hệt nhau
2. scatter chậm hơn chỉ thích hợp khi dữ liệu < few thousands

### Using plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')

# other markers
# 'o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'
```

### Using scatter 

```python
# get flower data from sklearn
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

# s: size, c: color
plt.scatter(features[0], features[1], alpha=0.2, s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

iris
```
